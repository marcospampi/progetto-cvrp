```python
import sys, os

sys.path.append(os.path.join(os.getcwd(),'..'))

from src.instance import Instance
from src.solution import Solution
from src.utils import show_solution
```

```python
INSTANCES_PATH = os.path.join(os.getcwd(),'..', 'instances')
INSTANCES_SOLUTION_PATH = os.path.join(INSTANCES_PATH, 'sol')
```

```python
import re

files = [
	file for file in os.listdir(INSTANCES_PATH) if re.search(r'\.vrp$', file) is not None
]

for file in files:
	title = file.split('.')[0]
	figsize = (12,8)
	instance = Instance.load_vrp(os.path.join(INSTANCES_PATH, file))
	solution = Solution.load_sol(os.path.join(INSTANCES_SOLUTION_PATH, file.replace('.vrp', '.sol')), instance, lenient = True)

	show_solution(instance, solution, title=title, figsize=figsize)
```
